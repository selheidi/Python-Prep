9) Imprimir los números primos existentes entre 0 y 30

```python
n=0
primo=True
while (n<31):
    for div in range (2,n):
        if (n%div==0):
            primo=False
        
    if (primo):
        print (n)
    
    else:
        primo=True
    n+=1
```
